## Change of Variables

Recall the formula for variable transformation as applied to probability density functions:

$p(y) = p(x)\left|\frac{dx}{dy}\right|$

Consider the function $b=\tan(\theta)$, which is sometimes used to reparametrize the slope of a line ($b$) with the angle the line makes in a plot or $x$ vs $y$ ($\theta$).

1. If $p(\theta)$ is uniform (proportional to a constant) on $-\frac{\pi}{2} < \theta < \frac{\pi}{2}$, work out $p(b)$.

2. Demonstrate that you're correct by generating a bunch of uniform random $\theta$'s, transforming each one to its corresponding $b$, and comparing a histogram of $b$ with your answer to (1).

3. Now try the reverse problem: find $p(\theta)$ if $p(b)$ is taken to be uniform (say over $-100<b<100$).

4. Compare your answer to (2) with a histogram of transformed samples from $p(b)$.

### My solution:

__1.__


> $b = \tan(\theta)$, so we take derivatives of both sides to find $db = \sec^2(\theta) d\theta)$.
>
> Then the change of variables uses $|\frac{d\theta}{db}| = \frac{1}{|1+\tan^2(\theta)|} = \frac{1}{1+b^2}$
>
> So we find the PDF of b on the interval $-\infty < b < \infty$: 
$$p(b) = p(\theta) |\frac{d\theta}{db}| = \frac{1}{1+b^2}$$ 

EDIT: forgot to normalize this to a probability distribution. The integral of $p(b)$ above gives $\pi$, so we have:
$$p(b) = \frac{1}{\pi(1+b^2)}$$ 


**2.**

```python
# import modules
import numpy as np
import matplotlib.pyplot as plt
```

```python
# generate array of random $\thetas$ and transform to b's
thetas = np.random.uniform(-np.pi/2, np.pi/2, size=2000)
bs = np.tan(thetas)

# generate points from answer of part 1
x = np.linspace(-20, 20, 1000)
probB = 1 / ((1 + x**2)*np.pi) #edited this, and changed the x scale to see the peak of the distribution more clearly.

# plot both results
plt.figure(figsize=(8, 5))
plt.hist(bs, bins=100, range=[-20,20], density=True, color='lightblue', label='histogram of samples')
plt.plot(x, probB, label='p(b)', color='indianred')
plt.legend()
plt.xlim([-20, 20])
plt.xlabel('b')
plt.show()
```

From the plot above, we find the theoretical (red) curve peaks in the same spot as the histogram from the randomly generated samples. Yay!

**3.** 

>This time we want the PDF of $\theta$ given $b$. If we do some algebra (invert the derivative and transform back to the right variable), we find that:
$$p(\theta) = p(b)|\frac{db}{d\theta}| = \sec^2(\theta)$$

EDIT: we should noramlize this. Because the given $b$ values are $-100<b<100$, we simply divide by $200$. 
$$p(\theta) = \frac{1}{200}\sec^2(\theta)$$

**4.**

```python
# generate array of random bs and transform to thetas
bs = np.random.uniform(-100, 100, size=2000)
thetas = np.arctan(bs)

# generate points from theoretical PDF
x = np.linspace(min(thetas), max(thetas), 1000)
probTheta = 1 / (200*np.cos(x)**2)

# plot both results
plt.figure(figsize=(8,5))
plt.hist(thetas, bins=100, density=True, color='lightblue', label='histogram of random $\\theta$ values')
plt.plot(x, probTheta, label='p($\\theta$)', color='indianred')
plt.legend()
plt.ylim([0, 20])
plt.xlabel('$\\theta$')
plt.show()
```

Again, the theoretical distribution matches the histogram.
